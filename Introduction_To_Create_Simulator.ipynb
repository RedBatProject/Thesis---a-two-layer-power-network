# Intro to networkx

```python
graph = top.create_nxgraph(Net)
```

```python
nx.draw(graph)
plt.show()
```

```python
alpha = 0.98
beta = 0.01
gamma = 1 - (alpha + beta)
n = len(graph.nodes())
scale_free = nx.scale_free_graph(n = n,alpha=alpha,beta=beta,gamma=gamma)
scale_free = nx.MultiGraph(scale_free)
nx.is_connected(scale_free)
```

```python
nx.draw(scale_free,with_labels=True)
plt.plot()
```

```python
print(f"radius: {nx.radius(graph)}")
print(f"diameter: {nx.diameter(graph)}") #maximum lenght
print(f"eccentricity: {nx.eccentricity(graph)}")
print(f"center: {nx.center(graph)}")
print(f"periphery: {nx.periphery(graph)}") # nodes with max lenght
print(f"density: {nx.density(graph)}")
```
